```python
import operator as operator


def solution(A):

    length = len(A)

    temp_dict = {}
    temp_list = []
    final_dict = {}
    counter = 0
    n = 0


    for x in range(length):
        # append first values into dictionary
        key = A[x]

        if key not in temp_dict.keys():
            counter = 0 # count first occurs
            temp_dict[key] = counter # append first values into dictionary
        else:
            y = x+1
            for y in range(length):
                # print(y)
                if(A[x] == A[y]):# matching values between new values & existing values in list ~ count++ by occurs times
                    counter += 1 # updating occurs time
                    temp_dict.update({key:counter}) # append updating values into dictionary
                    #print(A[x], ' - ', A[y])
                else:# no matching values between new values & existing values in list ~ adding new values into list and count by occurs times
                    new_key = A[y]
                    temp_dict[new_key] = 1

            # print(temp_dict)

            for key, value in temp_dict.items():
              temp_list.extend(temp_dict.items()) # because temp_dict.items() showing in array type ~ we have to using extend function


    z = len(temp_dict.keys())
    temp_list = temp_list[0:z]
    # print(temp_list)

    numbers = [item[0] for item in temp_list]
    occurs = [item[1] for item in temp_list]
    #print(numbers)
    #print(occurs)

    final_dict = dict(zip(numbers, occurs))
    #print(final_dict)

    
    temp = []
    #print(final_dict.items())
    
    
    for k,v in final_dict.items():        
        if final_dict[k] == final_dict[v]:
            a = k
            # if(final_dict.keys() == final_dict.values()):
            # solution = max(final_dict.items(), key=operator.itemgetter(1))[0] # get the final number with max occurs..
            # return solution
            temp.append(a)
            solution = max(final_dict.items(), key=operator.itemgetter(1))[0] # get the final number with max occurs..
            
            if solution > 0:
                return solution
            else:
                return 0
            
        else:
            return 0
        
        

```
