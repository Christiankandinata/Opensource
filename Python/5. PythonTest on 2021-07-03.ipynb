```python
maks = []
for i in range(0,len(keys)):
    if(keys[i] == values[i]):
        #print(keys[i])
        maks.append(keys[i])        
        
if(len(maks) > 0):
    print(max(maks))
elif(len(maks) == 0):
    print(0)
```

```python
## Final solution
def solution(a):
    maks = []
    
    counter = collections.Counter(a)
    keys = list(counter.keys())
    values = list(counter.values())
    
    for i in range(0,len(keys)):
        if(keys[i] == values[i]):
            #print(keys[i])
            maks.append(keys[i])        

    if(len(maks) > 0):
        print(max(maks))
    elif(len(maks) == 0):
        print(0)
```

```python
## Case 1
a = [3,8,2,3,3,2]
solution(a)
```

```python
## Case 2
a = [7,1,2,8,2]
solution(a)
```
