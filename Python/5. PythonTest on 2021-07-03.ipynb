```python
import collections
# a = [3,8,2,3,3,2]
# a = [7,1,2,8,2]
# a = [3,1,4,1,5]
a = [5,5,5,5,5]
```

```python
for i in range(0,len(a)):
    print(i)
```

```python
counter = collections.Counter(a)
print(counter)
```

```python
print(counter.values())
```

```python
print(counter.keys())
```

```python
keys = list(counter.keys())
values = list(counter.values())
```

```python
keys
```

```python
values
```

```python
if(keys == values):
    print(keys)
else:
    print(0)
```

```python
len(keys)
```

```python
maks = []
for i in range(0,len(keys)):
    if(keys[i] == values[i]):
        #print(keys[i])
        maks.append(keys[i])        
        
if(len(maks) > 0):
    print(max(maks))
elif(len(maks) == 0):
    print(0)
```
